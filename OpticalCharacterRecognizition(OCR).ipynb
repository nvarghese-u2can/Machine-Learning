```python
import keras_ocr
import matplotlib.pyplot as plt
```

Download the weights for the pre-trained detector and recognizer using the Pipeline method

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

Test images are stored in the /content directory within the Google Colab 

```python
images = [ keras_ocr.tools.read(img) for img in ['Img1.jpg','Img2.jpg']]
```

```python
print(f' Number of test images are : {len(images)}')
```

Test Images

```python
plt.figure(figsize = (10,20))
plt.imshow(images[0])
```

```python
plt.figure(figsize = (10,20))
plt.imshow(images[1])
```

Model Predictions

```python
predictions = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, predictions):
    keras_ocr.tools.drawAnnotations(image=image, 
                                    predictions=predictions, 
                                    ax=ax)
```
